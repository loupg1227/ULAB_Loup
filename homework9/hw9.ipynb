# 2 Plotting a Function
Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the following:
- An appropriate figure size (`figsize=(10,8)` is recommended)
- Labeled x and y axis
- A title
- Different colors and linestyles for each function
- A legend
- Your domain should be $x \in [0, 10]$. *Hint: Use `np.linspace` to set a range of values for your functions.*

Your function should return:
- A list of $f(x)$ values
- A list of $g(x)$ values.

```python
import numpy as np

def plot_of_x_squared_and_root_x():
    inputs = np.linspace(0, 10, 11)

    # Here I set up the domain inside the funtion
    
    square = inputs ** 2 
    root = np.sqrt(inputs) # I can either do a for loop to find my Y-values or I can just do this, and since this seems easier I choose this.

    plt.figure(figsize=(10,8))
    plt.gca().set_facecolor('black') # I quite like the black background color which is why I add this line. 

    plt.plot(inputs, square, color='white', linestyle="--", label="X squared") # The code for the square function which will have linestyle --, and be white.
    plt.plot(inputs, root, color = 'cyan', label="Square root of X") # The code for the square root function which will have default linestyle and cyan color. 
    plt.title(" X-squared and square root of X", fontdict={'fontsize': 20})
    plt.xlabel("X-label / Inputs", fontdict={'fontsize': 15}) 
    plt.ylabel("Y-label / Outputs", fontdict={'fontsize': 15}) # Both of the labels for the graphs seemed to be somewhat small so I make them larger.
    plt.legend() # Legend for both my functions, which is short because I've already labeled them in the lines above.
    plt.show()


    return square, root

plot_of_x_squared_and_root_x()
```

```python
import numpy as np

def plot_of_x_squared_and_root_x():
    inputs = np.linspace(0, 15, 16)

    # Here I set up the domain inside the funtion, and while it goes to 15, it will not plot to 15 because of the x and y limits
    
    square = inputs ** 2 
    root = np.sqrt(inputs) # I can either do a for loop to find my Y-values or I can just do this, and since this seems easier I choose this.

    # Making the figure themselves:
    
    plt.figure(figsize=(10,8))
    plt.gca().set_facecolor('black') # I quite like the black background color which is why I add this line. 
    plt.xlim(0,10)
    plt.ylim(0,100)

    # Plotting the functions I need to plot
    
    plt.plot(inputs, square, color='white', linestyle="--", label="X squared") # The code for the square function which will have linestyle --, and be white.
    plt.plot(inputs, root, color = 'cyan', label="Square root of X") # The code for the square root function which will have default linestyle and cyan color. 

    # Titlesand labels:
    
    plt.title(" X-squared and square root of X", fontdict={'fontsize': 20})
    plt.xlabel("X-label / Inputs", fontdict={'fontsize': 15}) 
    plt.ylabel("Y-label / Outputs", fontdict={'fontsize': 15}) # Both of the labels for the graphs seemed to be somewhat small so I make them larger.

    # Legends and lines:
    
    plt.axvline(x=5, color='red', linestyle='-', linewidth=2)
    plt.axhline(y=50, color='red', linestyle='-', linewidth=2)
    plt.legend(bbox_to_anchor=(1, 1), loc='upper left') # Legend for both my functions, which is short because I've already labeled them in the lines above.
    plt.show()


    return np.sum(square), np.sum(root)

print(plot_of_x_squared_and_root_x())
```
